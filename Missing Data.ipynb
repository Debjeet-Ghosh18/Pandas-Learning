```python
import pandas as pd
import numpy as np
```

**Finding Mising Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

**Removing Missing Data**

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3) # when we drop the values then we drop the values row wise
```

**Filling the missing Data**

```python
df.fillna(0) # When we fill the values that time we fill the values column wise
```

```python
fill_values = {"A":0,"B":100,"C":300,"D":400}
df.fillna(fill_values)
```

```python
df.fillna(df.mean())
```
