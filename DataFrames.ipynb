### Creating a DataFrame

```python
data = {
    'Name': ['Jhon','Anna','Peter','Linda'],
    'Age' : [28,34,29,42],
    'City': ['New York','Paris','Berlin','London'],
    'Salary': [65000,70000,82000,85000]
}
df = pd.DataFrame(data)
```

```python
df
```
